# 6. Seaborn
* matplotlib을 기반으로 다양한 색상 테마와 통계용 그래프 제공
* seaborn에서만 제공하는 통계용 그래프: countplot, boxplot, violinplot, heatmap
* DataFrame으로 되어있는 데이터를 시각화하는데 유용함.
* https://github.com/mwaskom/seaborn-data

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Seaborn에서 제공하는 datasets
df = sns.load_dataset('flights')
df.head()
```

## 6-1 Line Plot
* 파란색 부분은 데이터의 신뢰구간 표시

```python
plt.figure(figsize = (10,5))
sns.lineplot(x = 'year', y = 'passengers',data = df)
plt.show()
```

```python
sns.lineplot(x = df.year, y = df.passengers)
plt.show()
```

## 6-2. Scatter Plot


```python
sns.scatterplot(x = 'year', y = 'passengers',data = df)
plt.show()
```

```python
#겹치지 않게 보여주기
sns.swarmplot(x = 'year', y = 'passengers',data = df)
plt.show()
```

## 6-3. Bar Plot
* 검정색 막대(에러바)는 편차를 의미한다. 

```python
sns.barplot(x = 'year', y = 'passengers',data = df)
plt.show()
```

```python
df.loc[df.year == 1960,'passengers'].std()
```

```python
#hue를 지정하면 분류 한번 더 함.
plt.figure(figsize = (15,10))
sns.barplot(x = 'year', y = 'passengers', hue = 'month', data = df)
plt.show()
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.barplot(x = 'day',y = 'tip',data = tips)
plt.show()
```

```python
sns.barplot(x = 'day',y = 'tip',hue = 'time',data = tips)
plt.show()
```

## 6-4. Count Plot
* x축 데이터에 대해서 갯수를 세어줌

```python
sns.countplot(x = 'year' ,data = df)
plt.show()
```

```python
tips.head()
```

```python
sns.countplot(x = 'day',data = tips)
plt.show()
```

```python
sns.countplot(x = 'day', hue = 'sex',data = tips)
plt.show()
```

```python
sns.countplot(x = 'day', hue = 'time',data = tips)
plt.show()
```

```python
sns.countplot(x = 'day', hue = 'time', dodge = False, data = tips) #겹쳐서 나온다
plt.show()
```

## 6-5. Box Plot (Box-and-Whisker Plot)
* 최솟값, 최대값, 중앙값, 제1사분위수, 제3사분위수, 이상치(outlier)를 알 수 있음.
* outlier를 감지하기 위해서도 사용됨

```python
plt.figure(figsize = (10,5))
sns.boxplot(x = 'year', y = 'passengers',data = df)
plt.show()
```

```python
df.loc[df.year==1960].describe()
```

```python
# IQR (InterQuartile Range)= Q3-Q1
IQR =514.75 - 418.5
IQR
```

```python
#boxplot max 값
# Q3 + IQR * 1.5
514.75 + IQR * 1.5
```

```python
#boxplot min 값
# Q3 - IQR * 1.5
418.5 - IQR * 1.5
```

## 6-6. Violin Plot

```python
plt.figure(figsize = (10,5))
sns.violinplot(x = 'year',y = 'passengers',data = df[df.year == 1960])
plt.show()
```

## 6-7. heatmap
* 값의 변화를 색상표를 통해 알 수 있음

```python
flights = df.pivot_table(index = 'month', columns = 'year', values = 'passengers')
flights
```

```python
plt.figure(figsize = (10,8))
sns.heatmap(flights, annot = True, fmt = 'd') #정수로 출력! 실수는 f
plt.show()
```
