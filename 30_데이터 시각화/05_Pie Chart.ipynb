# 5. Pie Chart
* 백분율로 표시, 데이터릐 비율을 알아내는데 유용함


```python
import matplotlib.pyplot as plt
```

```python
data = [1,2,3,4]
plt.pie(data) #반 시계 방향
plt.show()
```

```python
labels = ['치킨','족발','팅글','케잌']
plt.pie(data,labels = labels)
plt.show()
```

```python
plt.pie(data,labels = labels, autopct = '%.1f%%')
plt.show()
```

```python
explode = [0.2,0,0,0]
plt.pie(data,labels = labels, autopct = '%.1f%%',explode = explode, shadow = True)
plt.show()
```

```python
colors = ['r','g','b','pink']
plt.pie(data,labels = labels, autopct = '%.1f%%',explode = explode, shadow = True, colors = colors)
plt.show()
```
