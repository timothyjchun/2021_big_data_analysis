## 문제) 성별 데이터를 남자=0, 여자 = 1로 바꾸기

```python
d.loc[d.성별 == "남자",'성별2'] = 0
d
```

```python
d.loc[d.성별 == "여자",'성별2'] = 1
d
```

```python
d['성별2'] = d['성별2'].astype(int)
d
```

## 11-1. Apply 사용하기

```python
#apply에 적용할 함수를 정의하기 (리턴값이 있어야함)
def male_or_female(x):
    if x == '남자':
        return 0
    else:
        return 1
```

```python
d['성별3'] = d['성별'].apply(male_or_female)
d
```

## 11-2. Map 사용하기

```python
my_map = {
    '남자':0,
    '여자':1
}
my_map
```

```python
d['성별4'] = d['성별'].map(my_map)
d
```

### 문제) 브랜드 평판지수 / 10000으로 나누기

```python
def brand_simplify(x):
    return round(x/10000)
```

```python
d['브랜드평판지수2'] = d['브랜드평판지수'].apply(brand_simplify)
d
```

### 문제2) 혈액형 데이터 바꾸기
* A형:0 B형:1 AB형:2 O형:3
* 혈액형2 컬럼에 넣기

```python
my_map = {
    "A":0,
    "B":1,
    "AB":2,
    "O":3
}
```

```python
d['혈액형2'] = d['혈액형'].map(my_map)
d
```
