# 13. Pandas 실습 (2)

```python
# numpy, pandas import하기
import numpy as np
import pandas as pd
```

## 13-1. 데이터 불러오기

```python
# 'http://bit.ly/student-score-2' URL로부터 csv 파일 불러오기 (첫번째 DataFrame)
d = pd.read_csv('http://bit.ly/student-score-2')
```

## 13-2. 결측값 확인하기

```python
# 결측값 갯수세기 (isnull() 이용)
d.isnull().sum()
```

```python
# 수학 점수에 결측값이 있는 행데이터 보기
d.loc[d['수학'].isnull()]
```

## 13-3. 결측값 처리하기

```python
# 수학 점수 평균구하기 (소숫점 첫째자리까지)
m = round(d['수학'].mean(),1)
m
```

```python
# 수학 점수의 결측값을 평균값으로 채우기
d['수학'].fillna(m,inplace = True)
d
```

```python
# 결측값 갯수세기 (isnull() 이용)
d.isnull().sum()
```

## 13-4. 자료형 변환하기

```python
# 국어, 수학 컬럼의 자료형을 float -> int로 변환하기
d['국어'] = d['국어'].astype(int)
d['수학'] = d['수학'].astype(int)
d
```

## 13-5. DataFrame 합치기

```python
# 'http://bit.ly/student-score-3' URL로부터 csv 파일 불러오기 (두번째 DataFrame)
d2 = pd.read_csv('http://bit.ly/student-score-3')
d2
```

```python
# 두개의 DataFrame을 열방향으로 합치기 (concat() 사용)
d_new = pd.concat([d,d2],axis = 1)
d_new.reset_index(drop = True)
d_new
```

## 13-6. DataFrame 데이터 삭제하기

```python
# 첫번째 DataFrame에서 지드래곤 삭제하기
d.drop(1)
```

```python
# 첫번째 DataFrame index reset하기 (reset_index() 이용하기)
d.reset_index(inplace = True,drop = True)
d
```

```python
# 두번째 DataFrame에서 뷔, 정국 삭제하기
d2.drop([3,5],inplace = True)
d2
```

```python
# 두번째 DataFrame index reset하기 (reset_index() 이용하기)
d2.reset_index(inplace = True, drop = True)
d2
```

## 13-7. DataFrame 병합하기

```python
# 두개의 DataFrame을 '이름' 기준으로 left 방식으로 merge하기
pd.merge(d,d2,on = '이름',how = 'left')
```

```python
# 두개의 DataFrame을 '이름' 기준으로 right 방식으로 merge하기
pd.merge(d,d2,on = '이름',how = 'right')
```

```python
# 두개의 DataFrame을 '이름' 기준으로 inner 방식으로 merge하기
pd.merge(d,d2,on = '이름',how = 'inner')
```

```python
# 두개의 DataFrame을 '이름' 기준으로 outer 방식으로 merge하기
pd.merge(d,d2,on = '이름',how = 'outer')
```

## 13-8. 데이터 변환하기
* 90점 이상: 'A', 80점 이상: 'B', 70점 이상: 'C', 그외: 'D' 부여하기
* 예) 국어점수가 95점인 경우 [국어등급] 컬럼에 'A' 넣기

```python
# Apply 함수 작성하기
def give_grade(x):
    if x >=90:
        return 'A'
    elif x<90 and x>=80:
        return 'B'
    elif x<80 and x>=70:
        return 'C'
    else:
        return 'D'
```

```python
# 첫번째 DataFrame에 국어, 수학 과목의 등급 부여하기
d['국어등급'] = d['국어'].apply(give_grade)
d['수학등급'] = d['수학'].apply(give_grade)
d
```

```python
# 국어등급 별 갯수 세기
d['국어등급'].value_counts()
```

```python
# 수학등급 별 갯수 세기
d['수학등급'].value_counts()
```

## 13-9. 데이터 그룹핑하기

```python
# pivot_table을 사용하여 반과 성별에 따른 국어, 수학 평균 구하기
pd.pivot_table(d,index = '반',columns = '성별',values = ['국어','수학'],aggfunc='mean')
```

```python
# groupby를 사용하여 반과 성별에 따른 국어, 수학 평균 구하기
pd.DataFrame(d.groupby(['반','성별'])[['국어','수학']].mean())
```
